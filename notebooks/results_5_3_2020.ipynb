```python
import math
import os
import IPython
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
vae_cifar_path = "/Users/leo/Documents/Tesla/alexnet-vae-no-padding/old_runs/logs_vanilla_vae_cifar10_27-02-2020/"
vae_deeper_cifar_path = "/Users/leo/Documents/Tesla/alexnet-vae-no-padding/old_runs/logs_deeper_vanilla_vae_cifar10_28-02-2020"
vae_even_deeper_cifar_path = "/Users/leo/Documents/Tesla/vae/master-thesis-leonard/old_runs/logs_even_deeper_vanilla_vae_cifar10_28-02-2020/"
columns = 17
```

## VAE with CIFAR10 - losses

```python
loss = []
r_loss = []
kl_loss = []
with open(os.path.join(vae_cifar_path,'2020-02-27.log')) as f:
    
    for line in f:
        line = f.readline()
        if "LossLoggingCallback.py" in line:
            splits = line.split(" ")
            loss.append(float(splits[7][:-1]))
            r_loss.append(float(splits[9][:-1]))
            kl_loss.append(float(splits[11][:-2]))
losses = [loss, r_loss, kl_loss]
loss_names = ['vae_loss','r_loss','kl_loss']
fig = plt.figure(figsize=(12,8))
ax = fig.gca()
arr = np.arange(len(loss))
for loss in losses:
    ax.plot(arr, loss)
    ax.legend(loss_names, loc='upper right')
plt.show(fig)
```

### highest level feature maps

```python
fm_path = os.path.join(vae_cifar_path, 'epoch_100/step_100/feature_map/')

num_samples = len(os.listdir(fm_path)) - 2

for sample_nr in range(num_samples):
    sample_path = os.path.join(fm_path, 'sample_{}'.format(sample_nr))
    num_maps = len(os.listdir(os.path.join(sample_path, 're_lu_4'))) + 1
    rows = math.ceil(num_maps/columns)
    
    fig=plt.figure(figsize=(columns*2, rows*2))
    fig.add_subplot(rows,columns,1)
    plt.imshow(Image.open(os.path.join(sample_path, 'original.jpg')))
    plt.axis('off')
    for i in range(2,num_maps):
        img = Image.open(os.path.join(sample_path,'re_lu_4', 'map_{}.png'.format(i-2)))
        img= np.array(img)
        fig.add_subplot(rows, columns, i)
        plt.imshow(img, cmap='gray')
        plt.axis('off')
    plt.show()
```
