```python
import math
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image


columns = 17
```

# Visualizations

## Classification - AlexNet - Layer 1 Kernels

```python
classification_path = '../results/alexnet_classification/epoch_10/step_739/'
num_maps = len(os.listdir(os.path.join(classification_path, 'layer1_kernels')))
rows = math.ceil(num_maps / columns)

fig=plt.figure(figsize=(columns, rows))
for i in range(1,num_maps +1):
    img = Image.open(os.path.join(classification_path,'layer1_kernels', 'map_{}.jpg'.format(i-1)))
    img= np.array(img)
    fig.add_subplot(rows, columns, i)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
plt.show()
```

## VAE - AlexNet $dim_z = 200$
